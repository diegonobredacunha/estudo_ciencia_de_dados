# Análise de dados com Python puro

```python
# O MÉTODO OPEN DO PYTHON LÊ UM ARQUIVO E JOGA PARA UMA VARIÁVEL
titanic_csv = open('/home/diego/workspace/python/estudo_ciencia_de_dados/Datasets/titanic.csv', 'r')
```

```python
# SEPARANDO O ARQUIVO EM LINHAS
linhas = titanic_csv.readlines()
```

```python
# PARA RETIRAR O DOCUMENTO DA MEMÓRIA
titanic_csv.close()
```

```python
# RETIRANDO A LINHA ZERO E USANDO COMO 
colunas = linhas[0].split(',')
```

```python
# DADOS RECEBERÁ O CONTEÚDO DA LINHA 1, SEPARADO POR VÍRGULAS, COMO FIZEMOS COM OS CABEÇALHOS
dados = linhas[1].split(',')
```

```python
# SEMPRE CONTAR O NÚMERO DE ELEMENTOS DA LISTA
print(len(colunas), len(dados))
```

```python
# NESSE CASO O NOME ESTÁ NO PADRÃO AMERICANO, AONDE VEM PRIMEIRO O ÚLTIMO NOME E DEPOIS O NOME E SOBRENOME.
# A FUNÇÃO TRATA O NOME, RETIRANDO A VÍRGULA E COLOCANDO NO PADRÃO BRASILEIRO
def tratar_nome(linha):
    import re
    match = re.compile(r'"(.*)(,)(\s.*)"') # REGEX, PEGA A STRING NOME
    return match.sub(r'"\3 \1"', linha) # PEGA O TERCEIRO BLOCO E COLOCA NA FRENTE DO PRIMEIRO. O SEGUNDO BLOCO, FOI EXLUÍDO, NO CASO A VÍRGULA
```

```python
# CRIANDO O DICIONÁRIO - O DICIONÁRIO TITANIC_DADOS ESTÁ RECEBENDO A COLUNA
# COMO CHAVE E UMA LISTA VAZIA COMO VALORES
# NA CRIAÇÃO DESSE DICIONÁRIO, CRIAREMOS AS CHAVES E OS VALORES COMO VAZIOS
titanic_dados = {}

for coluna in colunas:
    titanic_dados[coluna] = []

titanic_dados
```

```python
# AQUI FAREMOS A MESMA COISA, PORÉM VAMOS SEPARAR AS LINHAS PELA VÍRGULA E CONVERTER EM UMA LISTA
for dado in linhas[1:]:
    dado_tratado = tratar_nome(dado) # PRIMEIRO ELE TRATA O NOME E JOGA NA VARÁVEL
    dado_como_lista = dado_tratado.split(',') # DEPOIS ELE SEPARA CADA LINHA EM UMA LISTA
    print(dado_como_lista)
```

```python
dados_tratados = [] # CRIA UMA LISTA VAZIA

for dado in linhas[1:]: # FOR RODA A PARTIR DA LINHA 1, POIS A 0 É O TÍTULO DAS COLUNAS
    dado_tratado = tratar_nome(dado) # DADO_TRATADO = O RESULTADO DA FUNÇÃO TRATAR_NOME
    dado_como_lista = dado_tratado.split(',') # AQUI ELE QUEBRA LINHA E TRANSFORMA EM UMA LISTA
    dados_tratados.append(dado_como_lista) # ADICIONA AO FINAL DA LISTA MÃE AS LISTAS MENORES
```

```python
# 1 - PEGO CADA LINHA DA VARIÁVEL DADOS_TRATADOS E COLOCA EM PASSAGEIRO
# 2 - FAÇO UM FOR EM PASSAGEIRO E PEGO O ÍNDICE  E O RESTANTE DOS DADOS
# 3 - VOU NA VARIÁVEL COLUNA E PEGO A COLUNA NO RESPECTIVO INDICE

for passageiro in dados_tratados:
    for indice, dado in enumerate(passageiro): # ENUMERATE RETORNA INDICE E VALOR (ATRIBUIÇÃO MULTIPLA)
        coluna = colunas[indice]
        print(coluna, dado)
```

```python
# ADICIONA AS LISTAS AO DICIONÁRIO TITANIC_DADOS
for passageiro in dados_tratados:
    for indice, dado in enumerate(passageiro): # ENUMERATE RETORNA INDICE E VALOR (ATRIBUIÇÃO MULTIPLA)
        coluna = colunas[indice]
        titanic_dados[coluna].append(dado)
```

```python
# COLOCANDO UM VALOR NA IDADE VAZIA E AS CONVERTENDO EM NÚMERO
for indice, age in enumerate(titanic_dados['Age']):
    if titanic_dados['Age'][indice] != '':
        titanic_dados['Age'][indice] = float(age)
    else:
        titanic_dados['Age'][indice] = -1
```

```python
# COPIANDO OS VALORES DAS IDADE E COLOCANDO EM UMA LISTA FORA DO DICIONÁRIO
lista_idades = []
for idade in titanic_dados['Age']:
    if idade != -1:
        lista_idades.append(idade)
```

```python
# FUNÇÃO MÉDIA
def media(lista):
    soma_total = sum(lista)
    quantidade = len(lista)
    return soma_total / quantidade
```

```python
media_idades = media(lista_idades)
```
